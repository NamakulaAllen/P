```python
#poisson distribution
#it has two parameters
#iam-rate or known number of occurences e.g for above problems
#size-the shape of the returned array

#example: Generate a random 1x10 distribution for occurence 2

from numpy import random
x=random.poisson(lam=2, size=10)
print(x)
```

```python
#visualisation of poisson distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.displot(random.poisson(lam=2,size=1000), kde=False)
```

```python
#uniform distribution
#used to desribe probablity when  the event has equal chances
```

```python
#example : generate a 2x3 uniform distribution as a sample
from numpy import random
x=random.uniform(size=(2,3))
print(x)
```

```python
#Visualisation of Uniform Distribution

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.uniform(size=1000), hist=False)
plt.show()
```
